```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
positions = np.loadtxt('twogaussian_peakpos_peakutils.txt')
heights = np.loadtxt('twogaussian_peakheight_peakutils.txt')
widths = np.loadtxt('twogaussian_peakwidth_peakutils.txt')
```

```python
all_points_df = pd.DataFrame(all_points, columns=['Position', 'Height', 'Width', 'Time'])
all_points_df.describe()
```

```python
cluster = KMeans(n_clusters=2).fit(all_points[:,:-1])
cluster.labels_
```

```python
fignum = 1
fig = plt.figure(fignum, figsize=(7, 6))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=50, azim=123)
cluster
labels = cluster.labels_

ax.scatter(all_points[:, 0], all_points[:, 1], all_points[:, 2],
             c=labels.astype(np.float))

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Position?')
ax.set_ylabel('Height?')
ax.set_zlabel('Width?')
ax.dist = 12
```

```python
try1 = []
try2 = []

for i in range(1000):
    peak = cluster.predict([all_points[i][:-1]])
    if ( peak == 1 ):
        try1.append(all_points[i])
    elif ( peak == 0 ):
        try2.append(all_points[i])
    else:
        print('all_points[i] broke it!')

df = pd.DataFrame(try1, columns=['Position', 'Height', 'Width', 'Time'])
```
