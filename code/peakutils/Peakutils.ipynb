```python
# importing the libraries
import numpy as np
import peakutils
import syntheticdata
from peakutils.plot import plot as pplot
from matplotlib import pyplot
%matplotlib inline
```

# Find peak values of simple Gaussians model

__Find peak values of two simple Gaussians without noise__

```python
# generate data from two Gaussians without noise:
x = syntheticdata.nm
y = (syntheticdata.species_1 + syntheticdata.species_2)
pyplot.figure(figsize=(10,6))
pyplot.plot(x, y)
pyplot.title("Data without noise")
```

```python
# Getting a first estimate of the peaks
# By using peakutils.indexes, we can get the indexes of the peaks from the
# data. Due to the noise, it will be just a rough approximation.

indexes = peakutils.indexes(y, thres=0, min_dist=5)
print(indexes)
print(x[indexes], y[indexes])
pyplot.figure(figsize=(10,6))
pplot(x, y, indexes)
pyplot.title('First estimate')
```

__Find peak values of two simple Gaussians with noise__

```python
# generate some noisy data from two Gaussians:
x = syntheticdata.nm
z = (syntheticdata.species_1 + syntheticdata.species_2 + np.random.rand(x.size))
pyplot.figure(figsize=(10,6))
pyplot.plot(x, z)
pyplot.title("Data with noise")
```

```python
# Getting a first estimate of the peaks
# By using peakutils.indexes, we can get the indexes of the peaks from the
# data. Due to the noise, it will be just a rough approximation.

indexes = peakutils.indexes(z, thres=0, min_dist=5)
print(indexes)
print(x[indexes], z[indexes])
pyplot.figure(figsize=(10,6))
pplot(x, z, indexes)
pyplot.title('First estimate')
```

# Find peak values of complicated Gaussians model

```python
# generate numbers
x = syntheticdata.nm
# generate a two-gaussian mixture model with time decays but without spetral shift
m = syntheticdata.data_matrix_0
#generate a two-gaussian mixture model with time decays and spectral shift 
n = syntheticdata.data_matrix
```

```python
def index_matrix(nm_array,data_matrix):
    """
    generates peak values of each columns of data_matrix
    
    Args:
        nm_array: wavelength array
        data_matrix: a matrix that contains gaussian function at each time
    
    Returns:
        different peak values that each columns of data_matrix
        
    """
    for i in range(np.shape(data_matrix)[0]):
        indexs=peakutils.indexes(data_matrix[:,i], thres=0, min_dist=5)
        print (indexs)
    return
```

```python
index_matrix(x,m)
```

```python
index_matrix(x,n)
```

## Also, we can plot all the peaks, see the next function

```python
def peak_matrix(nm_array,data_matrix,range_number):
    """
    generates peak values of each columns of data_matrix and plot them in a diagram.
    
    Args:
        nm_array: wavelength array
        data_matrix: a matrix that contains gaussian function at each time
        range_number: the column numbers of data_matrix you want to use
    Returns:
        different peak values that each columns of data_matrix and diagram
        
    """
    pyplot.figure(figsize=(15,10))
    peak = np.zeros((range_number,2))
    for i in range(range_number):
        peak[i,:] =peakutils.indexes(data_matrix[:,i], thres=0, min_dist=5)
        pplot(x, m[:,i],peak[i,:].astype('int64'))
        pyplot.title('First estimate')
    return (peak)
```

```python
# generate a plot of 50 gaussian curves and return each columns of peak values
peak_matrix(x,m,50)
```
