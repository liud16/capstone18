```python
peaks_100 = np.array(idx_s.iloc[100])[:-1]
peaks_100_x = np.array([nm[int(peaks_100[0])], nm[int(peaks_100[1])]])
peaks_100_y = np.array([np.array(z_smooth)[int(peaks_100[0]), 100], np.array(z_smooth)[int(peaks_100[1]), 100]])
```

```python
peaks_100_n = np.array(idx.iloc[100])[:37]
peaks_100_x_n = np.empty(37)
peaks_100_y_n = np.empty_like(peaks_100_x_n)
for i in range(37):
    peaks_100_x_n[i] = nm[int(peaks_100_n[i])]
    peaks_100_y_n[i] = np.array(z_smooth)[int(peaks_100_n[i]), 100]
```

```python
peaks_100_x_n
```

```python
peaks_0 = np.array(idx_s.iloc[0])[0]
peaks_0_x = np.array([nm[int(peaks_0)]])
peaks_0_y = np.array([np.array(z_smooth)[int(peaks_0), 0]])
```

```python
peaks_450 = np.array(idx_s.iloc[450])[0]
peaks_450_x = np.array([nm[int(peaks_450)]])
peaks_450_y = np.array([np.array(z_smooth)[int(peaks_450), 450]])
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi = 300)

fs = 20
plt.xlabel('Wavelength (nm)', fontsize = fs, fontweight = 'bold')
plt.ylabel('Intensity', fontsize = fs, fontweight = 'bold')
plt.xlim(900,1400)
plt.xticks(fontsize = fs)
plt.yticks([])
plt.plot(nm, z[:, 0], color = 'orange', linewidth = 4)
plt.plot(nm, np.array(z_smooth)[:, 0], color = 'black', linewidth = 2)
plt.plot(peaks_0_x, peaks_0_y, 'o',  color = 'red', markersize = 10)


plt.plot(nm, z[:, 100], color = 'green', linewidth = 4)
plt.plot(nm, np.array(z_smooth)[:, 100], color = 'black', linewidth = 2)
plt.plot(peaks_100_x, peaks_100_y, 'o',  color = 'red', markersize = 10, label = 'After smoothing')
plt.plot(peaks_100_x_n, peaks_100_y_n, 'o',  color = 'red', markersize = 10, markerfacecolor = 'None', label = 'Before smoothing')

plt.plot(nm, z[:, 450], color = 'blue', linewidth = 4)
plt.plot(nm, np.array(z_smooth)[:, 450], color = 'black', linewidth = 2)
plt.plot(peaks_450_x, peaks_450_y, 'o',  color = 'red', markersize = 10)

plt.legend(fontsize = 18, edgecolor = 'black')
plt.show()
```
